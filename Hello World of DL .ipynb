# Load Data of MNIST dataset from keras

```
from tensorflow.keras.datasets import mnist
```

# Network Artchitecture

```
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```
network = Sequential()
```

```
network.add( layers.Dense(512, activation='relu', input_shape=(28*28,) ))
network.add(layers.Dense(10, activation='softmax'))
```

```
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
train_images = train_images.reshape(60000, 28*28)
train_images = train_images.astype('float32')/255
test_images = test_images.reshape(10000, 28*28)
test_images = test_images.astype('float32')/255
```

# This Step is called One-hot encoding

```
from tensorflow.keras.utils import to_categorical
train_lables = to_categorical(train_lables)
test_lables = to_categorical(test_lables)
```

```
train_lables[0]
```

Let's Train the Model

```
network.fit(train_images, train_lables, epochs=11, batch_size=128)
```

```
network.evaluate(test_images, test_lables)
```

The difference between training accuracy and tetsing accuracy is called overfitting
