```python
# feature2 has 0 values, it's useless for the data so will drop this
dataset.drop(columns='feature2', inplace=True)
```

```python
dataset.label.value_counts()
```

```python
# so here we will encode label column, b=1 and g=0
replace_values = {'label':{'g':0, 'b':1}}
```

```python
dataset = dataset.replace(replace_values)
```

```python
dataset = dataset.sample(frac=1)
```

```python
dataset
```

```python
y = dataset.label
x = dataset.drop(columns='label')
```

```python
x.head()
```

```python
y
```

```python
# now split tha data to train and test set
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size= 0.4 )
```

```python
X_train
```

```python
Y_train
```

```python
# Build The Model
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras import regularizers
```

```python
network = Sequential()
```

```python
network.add(layers.Dense(16, activation='relu', kernel_regularizer = regularizers.l2(0.001),  input_shape=(X_train.shape[1],)))
network.add(layers.Dense(8, activation='relu' , kernel_regularizer = regularizers.l2(0.001)))
network.add(layers.Dense(4, activation='relu' , kernel_regularizer = regularizers.l2(0.001)))
network.add(layers.Dense(1, activation='sigmoid'))
```

```python
network.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

```python
network.fit(X_train, Y_train, epochs=60, batch_size=8)
```

```python
prediction = network.predict(X_test)
```

```python
network.evaluate(X_test, Y_test)
```
